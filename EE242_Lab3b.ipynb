You will experiment with different FFT sizes and different plotting options. The description below assumes that you import numpy as np.

**A.** Use the np.fft.fft function to compute the FFT for the 100 ms horn signal, with an fft size of nfft=1024, which you can call **x_f**. Recall that the result of the FFT will be a vector that spans frequencies [0,$f_s$]. If this is a real-valued signal, then the first half of the FFT matters: [0,nfft/2]. **x_f** is not returned in the desired order so in order to get both the positive and negative frequencies, you need to use the np.fft.fftshift function to get **x_f2**. Create two different plots of the magnitude of result using **(np.abs(.))** in a 2x1 view: one with positive and negative frequencies and one with just positive frequencies. **B.** It is often the case that frequency content is plotted on a log scale. Plot the one-sided (positive frequency) magnitude using using a log scale.

**C.**  Changing the size of the FFT will change the frequency resolution, but it also changes the shape of the result a bit. Just as we saw with Gibbs phenomenon where increasing the number of Fourier series coefficients gave a high frequency ringing at sharp edges, increasing the FFT window will give a “ringing” effect for sharp peaks in frequency. To see this effect, compute the FFT using nfft=2048 and plot the log magnitude only using positive frequencies. Compare to your plot from Part B. The effect is easier to see on a log scale. 

```python
# Assignment 3 - Analyzing frequency content of a signal

# Part A - Using python FFT function
# Use the np.fft.ftt function to find the fourier domain representation of the horn signal
# Use the np.fft.fftshift to find the shifted fft
# Plot and see the positive and negative frequencies (remember to scale the FFT index by fs/nfft and magnitude by fs*signal_duration)
    # Seperate out the positive part into another vector
    # Plot the seperated signal
def analyze_frequency(file):
    sample_rate, audio_array = wavfile.read(file)
    
    segment_duration = 0.1  # 100 ms
    segment = audio_array[:int(sample_rate * segment_duration)]
    
    nfft_1024 = 1024

    x_f = np.fft.fft(segment, n=nfft_1024)
    x_f2 = np.fft.fftshift(x_f)
    freq = np.fft.fftfreq(nfft_1024, d=1/sample_rate)
    freq_shifted = np.fft.fftshift(freq)   

# Part B - Plot the FFT across frequency
    plt.figure(figsize=(10, 8))
    
    plt.subplot(2, 1, 1)
    plt.plot(freq_shifted, np.abs(x_f2) / len(segment))
    plt.title("Positive and Negative Frequencies FFT Magnitude")
    plt.xlabel("Frequency")
    plt.ylabel("Magnitude")
    
# Repeat the process for plotting the positive frequencies
    plt.subplot(2, 1, 2)
    plt.semilogy(freq[:nfft_1024 // 2], np.abs(x_f[:nfft_1024 // 2]) / len(segment))
    plt.title("Positive Frequencies FFT Magnitude")
    plt.xlabel("Frequency")
    plt.ylabel("Magnitude")
    
    plt.tight_layout()
    
# Part C - Horn Signal Ringing effect.
# Do the same things as Part A but with nfft = 2048
    nfft_2048 = 2048
    x_f_2048 = np.fft.fft(segment, n=nfft_2048)
    freq_2048 = np.fft.fftfreq(nfft_2048, d=1/sample_rate)
    
    mag_1024 = np.abs(x_f[:nfft_1024 // 2]) / len(segment)
    mag_2048 = np.abs(x_f_2048[:nfft_2048 // 2]) / len(segment)
    
    plt.figure(figsize=(10, 6))
    plt.semilogy(freq[:nfft_1024 // 2], mag_1024, label="1024")
    plt.semilogy(freq_2048[:nfft_2048 // 2], mag_2048, label="2048")
    plt.title("FFT Comparison (Log)")
    plt.xlabel("Frequency")
    plt.ylabel("Magnitude (Log)")
    plt.legend()
    plt.show()
# Plot the the signal (positive frequencies only) and observe the additional ringing -- compare with Part B

analyze_frequency("horn11short.wav")
```
